```python
preco = float(input(' Qual o valor da etiqueta do produto? R$'))

print(' \n Condições de pagamento: '
      ' \n 1- À vista em dinheiro ou cheque, recebe 10% de desconto'
      ' \n 2- À vista no cartão de crédito, recebe 5% de desconto'
      ' \n 3- Em 2 vezes, preço normal de etiqueta sem juros'
      ' \n 4- Em 3 vezes, preço normal de etiqueta mais juros de 10%')

condicao = int(input('\n Digite o código da opção de pagamento: '))

print('\n Preço de etiqueta do produto: ', preco)

if condicao == 1:
  preco = preco*0.9
elif condicao == 2:
  preco = preco*0.95
elif condicao == 3:
  preco = preco
elif condicao == 4:
  preco = preco*1.1
else:
  print(' Condição de pagamento inválida. ')

print(' Preço a pagar: R$', preco)

```

```python
print(' * QUADRO OPERAÇÕES ENTRE DOIS VALORES * \n')

n = int(input(' Informe o número de cálculos que deseja efetuar: \n'))
while n <= 0:
  print(' Opção inválida')
  op = int(input(' Informe o número de cálculos que deseja efetuar: '))

i = 1

while i <= n:
  print('\n MENU ')
  print(' (1) Adição')
  print(' (2) Subtração')
  print(' (3) Multiplicação')
  print(' (4) Divisão entre números reais')
  print(' (5) Divisão entre números inteiros')

  op = int(input('\n Informe o valor da opção: '))

  while op < 1 or op > 5:
    print(' Opção inválida')
    op = int(input(' Informe o valor da opção: '))

  print('\n Opção escolhida com sucesso \n')

  n1 = float(input(' Informe o primeiro valor: '))
  n2 = float(input(' Informe o segundo valor: '))

  if op == 1:
    res = n1+n2
    print('\n O valor da soma é: ', res)
  elif op == 2:
    res = n1-n2
    print('\n O valor da subtração é: ', res)
  elif op == 3:
    res = n1*n2
    print('\n O valor da multiplicação é: ', res)
  elif op == 4:
    if n2 == 0:
      print('\n Não é possível fazer a divisão por 0')
    else:
      res = n1/n2
      print('\n O valor da multiplicação é: ', res)
  else:
    if n2 == 0:
      print('\n Não é possível fazer a divisão por 0')
    else:
      res = n1//n2
      print('\n O valor da multiplicação é: ', res)
  i = i+1
```

```python
print(' *PROGRAMA DE GASTOS HOSPITALARES* \n')

n = int(input(' Quantos pacientes você quer calcular? '))
while n <= 0:
  print('Número inválido: Menor ou igual a zero.')
  n = int(input(' Quantos pacientes você quer calcular? '))

i = 1

while i <= n:

  print('\n Responda referente ao paciente número: ', n)

  d = int(input('\n Quantos dias o paciente ficou no hospital? '))
  while d <= 0:
    print('Número inválido: Menor ou igual a zero.')
    d = int(input(' Quantos dias o paciente ficou no hospital? '))

  print('\n Tipos de quartos:'
        ' (1) - Particular'
        ' (2) - Semi-particular'
        ' (3) - Coletivo')
  q = int(input(' Qual tipo de quarto o paciente ficou? '))
  while q < 1 or q > 3:
    print(' Opção inválida: Número não correspondente ao menu mencionado. \n')
    print('\n Tipos de quartos:'
        '\n (1) - Particular'
        '\n (2) - Semi-particular'
        '\n (3) - Coletivo')
    q = int(input('\n Qual tipo de quarto o paciente ficou? '))

  w = int(input(' O paciente utilizou WIFI? (1) Sim - (2) Não'))
  while w != 1 and w != 2:
    print(' Opção inválida: Somente disponível (1) Sim - (2) Não')
    w = int(input(' O paciente utilizou WIFI? (1) Sim - (2) Não'))

  t = int(input(' O paciente utilizou TV a cabo? (1) Sim - (2) Não'))
  while t != 1 and t != 2:
    print(' Opção inválida: Somente disponível (1) Sim - (2) Não')
    t = int(input(' O paciente utilizou TV a cabo? (1) Sim - (2) Não'))

  if q == 1:
    qt = ('Particular')
  elif q == 2:
    qt = ('Semi-particular')
  else:
    qt = ('Coletivo')

  if q == 1:
    qp = 360.00
  elif q == 2:
    qp = 210.00
  else:
    qp = 185.00

  if w == 1:
    wp = 3.00
  else:
    wp = 0.00

  if t == 1:
    tp = 4.00
  else:
    tp = 0.00

  print('------------------------------------------------------\n'
        ' Relatório de gastos do paciente ', i,
        '\n Números de dias no hospital: ', d,
        '\n Tipo de quarto: .............', qt,
        '\n Diárias: ....................', qp*d,
        '\n WIFI: .......................', wp,
        '\n TV a cabo: ..................', tp,
        '\n Total: ......................', qp*d + wp + tp)

  i = i + 1
```

```python
print(' *CONVERSOR DE METRO PARA OUTRAS UNIDADES DE MEDIDAS* ')

n = int(input(' Quantos medidas você vai converter? '))
while n <= 0:
  print('Número inválido: Menor ou igual a zero.')
  n = int(input(' Quantos medidas você vai converter? '))

i = 1

while i <= n:

  print('\n Bem-vindo ao conversor de medidas! ')

  m = float(input('\n Por favor, insira a sua medida em Metros: '))
  while m <= 0:
    print('Número inválido: Menor ou igual a zero.')
    m = float(input(' Insira a medida em Metros: '))

  print('\n Para qual unidade você deseja converter sua medida? '
        '\n (1) - Milímetros(mm)'
        '\n (2) - Centímetros(cm)'
        '\n (3) - Quilômetros(km)'
        '\n (4) - Polegadas(in)'
        '\n (5) - Pés(ft)'
        '\n (6) - Jarda(yd)'
        '\n (7) - Milha(mi)')
  op = int(input())

  while op < 1 or op > 7:
    print('Opção inválida: Escolha uma opção do respectivo menu.')
    print('\n Para qual unidade você deseja converter sua medida? '
        '\n (1) - Milímetros(mm)'
        '\n (2) - Centímetros(cm)'
        '\n (3) - Quilômetros(km)'
        '\n (4) - Polegadas(in)'
        '\n (5) - Pés(ft)'
        '\n (6) - Jarda(yd)'
        '\n (7) - Milha(mi)')
    op = int(input())

  if op == 1:
    res = m*1000
    print('A medida convertida é: ', res,' milímetros.')
  elif op == 2:
    res = m*100
    print('A medida convertida é: ', res,' centímetros.')
  elif op == 3:
    res = m*0.001
    print('A medida convertida é: ', res,' quilômetros.')
  elif op == 4:
    res = m*39.3701
    print('A medida convertida é: ', res,' polegadas.')
  elif op == 5:
    res = m*3.28084
    print('A medida convertida é: ', res,' pés.')
  elif op == 6:
    res = m*1.09361
    print('A medida convertida é: ', res,' jardas.')
  else:
    res = m*0.000621371
    print('A medida convertida é: ', res,' milhas.')

  i = i + 1
```
